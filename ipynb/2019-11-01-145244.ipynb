Title:本栖湖キャンプ

以前は日本をバイクで周りつつそのへんにテント張って生活している期間が長かったが、最近はもうそういう生活はいいかな、と思うようになった。疲れるし。

だがたまにはテントに泊まるのはやってもいいな、という気がする。
以前のホームレスみたいなのじゃなくて、もうちょっとちゃんとしたレジャーみたいなの。
ちょうどゆるキャンも見たし、秋に富士五湖のあたりにキャンプに行こう、となった。

泊まる場所は幾つかググったが、チェックインが早いとたどり着けない可能性があるしなぁ、と思ったところ、本栖湖のキャンプ場は予約不可なので行ってみてダメなら他を考える、というスタイルなので、まずは本栖湖のキャンプ場に行ってみたら泊まれた。
ボッチは本栖湖。島りんも言ってた。

バイクだと一泊1800円くらい。ソロキャン的には高いが、レジャーとして短期泊まるならなんでも無い。
一泊で十分な距離なんだが、翌朝すぐ撤収するのがかったるいので二泊する事に。今回はキャンプが目的なので、ツーリングはほどほどで。

キャンプ場はびっくりするほど混んでたが、テントを張れない程でもなく、張ってしまえば周りはそんな気にならない。
富士山もキレイだし、まぁいいか。

テント設営終わったら15:30くらい。
最初はどこか行こうかと思ったが、どこかといってもどうせ他の湖を回る、とかになるので、それなら本栖湖歩けばいいんじゃないか？という気がして本栖湖を歩いてみる。

本栖湖自体は結構ガラガラで釣り人は結構居るが道路散歩してる分にはほとんど人も居ないし、結構いいな。

その後夜になったのでお約束のカレーヌードルを食べる。お湯を沸かしたら、せっかくなのでそのままコーヒーも入れたりしてまったりする。

一人用テントしか張れ無さそうな所に張ったので周りは人も居なくて、夜になったら意外と良いな。
秋は蚊が居ないのもいいね。

### 2日目

朝8:30に目覚め。よく寝たな。
明け方はやや寒かったが、シェラフカバーは結局出さなかった。

明日は富士山をまわって帰りたいから、今日は甲府側に行くかな。といっても町中行っても仕方ないが。せっかくなのでフルーツ公園行くか。
この辺は所沢時代に何度も通ってるが、行ったことは無いはず。

フルーツ公園は行くまでの道がなかなかツーリング的に良かった。フルーツラインという名前がついてるらしい。
公園は悪くないが普通。

そのあとは温泉入って四尾連湖よるか？と思ったが、温泉入ったら夕方15時くらいだったので無理せず帰る事にする。
温泉は一瞬ほったらかし温泉も考えたが、場所的にもうちょっとキャンプ地に近い方がいいな、と思い、前回も来たみたまの湯。

風呂はなかなか良いのだが、そこで体重計に乗ったらめっさ増えてる！
家の体重計では数日前はそんなでも無かったが…

今回のツーリングで増えた分には戻ってから落とせば良いが、体重計がくるってるとかだったら嫌だなぁ。
長い事太ってたら落とすの大変なので。
まぁ帰ってみないと分からないのでこの件は保留だな。

キャンプ場にもどると、富士山は雲がかかってて見えない。
まぁいいか、と暇つぶしに濡れた木で火をおこせないか少しチャレンジして挫折してたら、ふと見上げると富士山が綺麗に出てる。せっかくなのでコーヒーを入れて飲みつつ富士山を見る。至福。

そうこうしてるうちに現在18:40で大分夜になった。昨晩よりだいぶ寒いな。明け方の寒さ耐えられるか？
一応もう二段階くらい寒さ対策を持ってきてはいるので、昨晩より寒い程度は想定内なのだが…

寒いおかげで、今日は星が良く見えるな。周りの夕飯が終わったくらいで一度見に出るかな。
本当は夜遅い方がいいが、寝たら起きない気がするので。

21時くらいには霧が出てきてしまった。今日はダメか。

### 夜中は冬の星座

夜中トイレに目が覚めたら空が完全に晴れてて冬の星が良く見えた。
もう冬かぁ。
冬は車持ってないのでそんなに星を見る機会は無いから秋や夏ほどは見慣れてないのと、今シーズン初なので星座早見表を見ながら星をたどる。
大犬座、牡牛座、双子座などの定番の星座は良く見えた。

本栖湖は山に囲まれているので見える範囲はそんなに大きくない。暗さはかなりのものだが。富士山の左側が少し明るいが、あれは富士宮とかの方か？

流れ星は2つほど見れた。今晩は人工衛星が見当たらなかったな。
なかなか堪能した。よしよし。

### 3日目

明け方は寒くて百均のアルミシートを導入。これは耐久性は無いが防寒性はめっちゃ高くて以前は良くお世話になってた。
今回も使い捨てのつもりで買っておいたが、買っておいてよかったかな。
これを使えば全然平気な程度の寒さだった。

無くても一時間くらい耐えれば暖かくなるので耐えられたとは思うが。
今回は季節的にはちょうどよかったな。11月後半だともう自分の装備では無理そう。

帰りは朝霧から富士スカイラインを通って帰ろう。富士スカイラインはいいね。
思ったより紅葉がまだだったが、多少は色づいてたし富士山も綺麗で気分爽快。
この道はいいよなぁ。

その後なんかウィンカーの調子が悪くなったのでまっすぐ帰る事に。
帰って開けてみたらバルブがとれかかってたので入れ直した。
このくらいなら出先で直せたなぁ。

とにかくお疲れさまでした。
天気も気候も良くて、なかなか楽しかった。
疲れ具合的にもこのくらいが丁度良いな。
一日走るを2日とかやると疲れてしまうので。

Title:茂木〜大子をツーリング

### 5月22日

引っ越す前に茨城を回っておこう、という事で今回は北側を回る。天気も良いしね。

常磐道で谷和原あたりまで出て、以後は下道で北上しようかな。

12:36 加平パーキング。積荷のチェックとかの為に少し寄る。こんな近いパーキング来た事無かったが。

14:46 道の駅にのみや。
いちごが300円一パックで美味しそうだったので買う。なかなか美味。

谷和原からのしたみちはなかなか気分良く走れた。
こっち側はこんなすぐ田舎道っぽくなるのね。
埼玉、千葉側はもっと住宅地が続く感じだが。

16時ころ 益子のあたりで休憩。

18時ころ常陸大宮の道の駅で夕飯。
茂木のあたりの道はとても良かった。
この辺はツーリング向きだなぁ。

ただこのさきは観光地で止まる場所が無さそうなので、常陸那珂に戻って一泊。

### 5月23日

昨晩はシェアハウスの同室の人の実家に泊めてもらった。10時に出発。

12時ころ 奥久慈大子の道の駅。
ちょっと早いが、ここからは店が無さそうなのでここで飯。

少し散歩をする。夏が来るね〜。

14時 馬頭のコンビニでトイレ休憩。奥久慈からここまでの道はなかなか良かった。

14:30 きれつがわの道の駅の前のスーパーで作戦会議。
ここまでが今回の予定のルートで、ここからは考えてなかったので。
新4号を外環まで降りてそこから千葉側に行く事に決定。

ついでに100均で装備を整える。
バイクには、これまで押すと一瞬赤く光るタイプの腕時計をつけてたが昼の日の下でまったく見えないので普通の液晶タイプに変更。
あのタイプは使えないという事がわかった。

16:10 道の駅しもつけで休憩。新四号は何も面白くはないが快適だな。
これなら高速じゃなくてもいいね。

と思ったが圏央道とぶつかったあたりから信号多くてだるいな。

24:30 千葉によって冬服を閉まったりして、家に着いたのはこの時間。お疲れ様でした。
外環はめっちゃ便利になったね！
